```python
meal_cost = float(input("Enter the cost of the meal: $"))

tax_rate = 0.13

tax_amount = meal_cost * tax_rate
tip_amount = meal_cost * 0.18
total_amount = meal_cost + tax_amount + tip_amount

print("Tax: ${:.2f}".format(tax_amount))
print("Tip: ${:.2f}".format(tip_amount))
print("Total: ${:.2f}".format(total_amount))
```

```python
sqperacre = 43560

length = float(input("Enter the length of the field in feet: "))
width = float(input("Enter the width of the field in feet: "))
area = (length * width) / sqperacre
print("The length of the field in feet is {length}, ""the width of the field in feet is {width}, ""and the area of the field in acres is {area}".format(length=length, width=width, area=area))
```
